Comando para apagar todas as colunas que todos os valores são nulos, pois elas não tem dados a serem analisados

Nesse dataset não temos colunas totalmente nulas, mas fica de exemplo

```python
dados.dropna(axis=1, how='all',inplace=True) #dropo colunas(axis=1), onde todos os valores são nulos(how='all)
```

Podemos alterar os dados nulos, invés de simplesmente dropar as linhas que tem valores nulos. Isso é MUITO importante, pois certas análises não permitem droparmos linhas, ou também simplesmente não há necessidade de dropar a linha toda quando podemos simplesmente preencher os dados para realizarmos as análises.

```python
dados.loser_entry.unique()
```

```python
dados.loser_entry.fillna('No Entry', inplace=True) #passo o valor a ser preenchido na variável nula e falo pra alterar de fato a variável
```

```python
dados.loser_entry.unique()
```

Alterar mais de um atributo de uma vez, com condições diferentes.

Para alterar mais de uma variável de uma vez, podemos usar python dicts e passar as alterações no estilo chave:valor com a coluna e os dados que deverão ser preenchidos nos valores nulos

```python
dados.w_ace.unique()
```

```python
dados.loser_ht.unique()
```

```python
dados.fillna(value={'w_ace':0,'loser_ht':dados.loser_ht.mean()},inplace=True) #passo o valor 0 para nulos em w_ace, e a média de altura dos perdedores que não estão nulos para loser_ht
```

```python
dados.w_ace.unique()
```

```python
dados.loser_ht.unique()
```

Projetando dados anterior em valores nulos usando o method ffill

```python
dados.minutes[2650:2700]
```

```python
dados.minutes = dados.minutes.fillna(method='ffill')
```

```python
year=[] #Coletando dados de cada coluna e criando arrays. Vou deixar a coluna de % de crescimento de fora pois eu mesmo vou criar ela.
avg_stock_price=[]
year_open=[]
year_high=[]
year_low=[]
year_close=[]

linhas = table.find_all('tr')
for linha in linhas:
  dados=linha.find_all('td')
  year.append(dados[0].text)
  avg_stock_price.append(dados[1].text)
  year_open.append(dados[2].text)
  year_high.append(dados[3].text)
  year_low.append(dados[4].text)
  year_close.append(dados[5].text)
```

```python
#criando colunas do dataframe
apple_stock = apple_stock.assign(avg_stock_price=avg_stock_price,
year_open=year_open,
year_high=year_high,
year_low=year_low,
year_close=year_close)
```

```python
apple_stock.sort_values(by=['year'], ignore_index=True, inplace=True) #organizando os anos e os indíces e alterando a variável original
```

Deslocamento de dados. com o .shift eu desloco dados para baixo e passo a quantidade de indíces que quero deslocar(nesse caso =1). Para ilustrar isso, criei uma coluna com o stock price do ano anterior

```python
apple_stock['last_year_close'] = apple_stock.year_close.shift(1)
```

```python
apple_stock
```

Nova coluna com a diferença do stock price de um ano para o outro

```python
apple_stock = apple_stock.astype(float)
```

```python
apple_stock['alteracao_%'] = ((apple_stock['year_close']/apple_stock['last_year_close']-1)*100)
```

```python
apple_stock.year = apple_stock.year.astype(int)
```
