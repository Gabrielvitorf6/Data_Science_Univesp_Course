## ***Series***

São objetos do tipo array unidimensional com uma sequência de valores e um array associado com labels(rótulos) de dados, o qual é chamado de "índice".

As Series são podem ter um tipo de dados pelo fato de ser unidimensional

Observamos aqui os índices ao lado dos valores, como vimos antes na aula

```python
serie1 = pd.Series([1,2,-10,-15])
```

```python
serie1
```

Vemos os valores da Serie com o método .values

```python
serie1.values
```

Vemos o método .index, que mostra o índice do começo da Serie(0), o fim(4) e o passo a passo da progressão do índice(De um em um).

Sempre que fazemos valores, o primeiro índice é inclusivo e o segundo é exclusivo. O Inclusivo: Índice que é incluso na fatia/indexação e o Exclusivo é o ìndice onde o índice para de ser contado(Com o próprio índice declarado já sendo exxcluído)

```python
serie1.index
```

Acessando valores por índice

```python
serie1[0]
```

Posso atribuir intervalos de outras Series para uma nova Series passando os índices que quero trazer, e retornando o índice e seu valor original(Nesse caso letras invés de números) e seus respectivos valores

```python
seriesintervalo=pd.Series(series2[0:3])
```

```python
seriesintervalo
```

Posso também atribuir valores de intervalos de uma variável a outra variável, e depois mudar os seus índices(Aqueles que vem automático) para os quais eu quiser

Aqui vamos alterar a variável series2

```python
series2 #Original
```

```python
index_values = [1,2,3]
series2alterada=pd.Series(series2[0:3])
series2alterada = series2alterada.set_axis(index_values)
"""Criamos uma variável chamada index_values com os valores que virarão os índices(1,2,3)
 da Series series2alterada(que está recebendo o intervalo de 0 a 3 da Series series2, junto de seus respectivos valores e índices originais).
Depois de criarmos a variável e a Series, usamos o método set_axis e passamos como parâmetro a variável index_values, e assim alteramos os índices que estavam na Series."""
```

```python
series2alterada
```

# *DATAFRAMES*

Os dataframes são arrays multidimensionais(Semelhantes a tabelas do Excel), que são formados ppor várias Series. Os dataframes são comumente chamados de df, e , ao contrário das Series, um dataframe pode ter VÁRIAS colunas em que cada uma delas tem um tipo de dado específico.

No dataframe temos índices, labels e valores. As labels(rótulos) são os nomes das colunas e as chaves para indexar valores, exemplo: '**estado**':[***]

```python
dados={'estado':['SP','MG','PR','GO','PR'],'ano':[2010,2011,2012,2013,2014],'pop':[16.9,87.9,52.2,56.4,22.4]}
```

Até aqui, temos a estrutura de um dicionário Python(dict). Para de fato criarmos um DataFrame do pandas, usamos o método .DataFrame e passamos como parâmetro os dados

```python
df1=pd.DataFrame(dados)
```

```python
df1
```

**Posso excluir colunas específicas do DF**


  Podemos excluir valores de duas formas, usando o comando .drop(do pandas) ou del(keyword do próprio Python).
A keyword del TEM BROADCAST, então caso usarmos ela para excluir valores sem usar o método .copy antes de atribuir valores a uma variável nova,
 a variável original será alterada

```python
df1.drop('pop', axis='columns') #especificamos a coluna que queremos dropar e qual o axis dela(axis=1 coluna e axis=)
#.drop sem broadcast
```

```python
df1 #Variável original intacta
```

```python
#Caso nós queiramos, podemos alterar o valor de uma variável por meio do .drop usando "inplace=True" como parâmetro na chamada do método
#Vale lembrar, que se caso NÃO usarmos o método .copy na criação da variável, a variável original sofrerá broadcast também
```

```python
df1copiaparainplace=df1.copy()
```

```python
df1copiaparainplace.drop('pop', axis = 1, inplace=True)
```

```python
df1copiaparainplace #Coluna da variável de FATO foi excluída
```

```python
df1cópiaparaexcluir=df1.copy() #Criando variável nova para demonstrar o broadcast
df1colunasexcluidas=df1cópiaparaexcluir
```

```python
df1cópiaparaexcluir #Variável original
```

```python
del df1colunasexcluidas['pop']
```

```python
df1colunasexcluidas
```

```python
df1cópiaparaexcluir #Variável original que sofreu broadcast
```

**Posso excluir linhas também**

```python
df1cópiaparaexcluir.drop([0,1]) #Aqui também a variável de fato não é excluída, e precisa do inplace=True para ser modificada internamente.
```
