We need to open our .txt file(info about brazilian national soccer team) and load this file to collab(folder button in page left side).

**Note**:When we close this colab environment, we going to lose this archive and we will need to load it again.

```python
arquivo=open('Selecao_brasileira.txt', 'r') #We use python function 'open' that
#open files, and after that, specify that we using read mode('r')
```

```python
conteudo = arquivo.readlines() #Use .readlines to read all lines from our file
arquivo.close() #close the file .arquivo to optimize memory, since we dont need it anymore
```

```python
print(conteudo)
print('---')
for linha in conteudo:
  print(linha)
```

Replacing 'FIFA' with 'Federao Internacional de Futebol'

```python
x='' #Create an empty variable to receive our data and get manipulated.
for linha in conteudo:
  if linha != '': #If our line is not empty
    x+=linha

print(x)
x=x.replace('FIFA', 'Federao Internacional de Futebol') #Then, we replace where 'FIFA' by our string
print('------------')
print(x)

#At this point, we already realized how almost everytime that we need to modify strings, values or index order,
#we NEED to create a empty variable,store the values that will be changed, and then, change the values
#using methods like .replace
```

Now we need to write our files. The process, is basically the same to read files, but with write values.

```python
arquivo_novo=open('Selecao_brasileira_novo.txt', 'w') #Specify the name of the file that we going to open,
#then, we specify the MODE that we going to use, in this case, write

arquivo_novo.writelines(x) #Use function .writelines in new variable,and pass a  variable as parameter (x with FIFA replaced)
arquivo_novo.close() #At this point, we already created a new file with written data from x, and it's
#already in folder button, in page left side
```

#**Working with CSV files**

Now, we using this dataset from Tesla stock dataset:image.png

```python
#import csv
with open('TSLA.csv', 'r') as file: #using open function and returning a variable 'file'
  leitor=csv.reader(file, delimiter=',') #we use csv reader and specify what's the delimiter(csv always uses delimiter)
  for linha in leitor: #On each loop repeat(each line from dataset), we store data from line 1, line 5 and line 7
    data=linha[0]
    fechamento=linha[4]
    volume=linha[6]
    print('Data:',data,'   Fechamento:',fechamento,'   Volume:',volume)

    #Note that this time, we don't need to close the file, since we use 'with open as'
```

**Working with CSV file using Pandas**

Notice that we get the data directly from web, since we provide the link from web page with raw file(This wouldn't work if we tried to retrieve the data directly from github page)

```python
arquivo='https://raw.githubusercontent.com/JeffSackmann/tennis_atp/refs/heads/master/atp_matches_2021.csv' #link from raw file
arquivo_lido=pd.read_csv(arquivo)
arquivo_lido #dataframe
```

We can plot the information from CSV using Pandas.

Here we're going use info about winner's dexterity(This dataset provides this info but dues to it's size, we cannot see this column when printing the dataset)

```python
arquivo_lido.winner_hand.value_counts().sort_values() #Acess data from column 'winner_hand -> count the values
#-> sort the values
```

We can also display a graph:

```python
arquivo_lido.winner_hand.value_counts().sort_values().plot(kind='bar')

#An easy visualization show how most of tennis players are right-handed
```

#**Working with JSON files**

Getting data from api https://randomuser.me/api that generates random data from a fictional person

```python
import json
import requests
```

```python
resposta=requests.get('https://randomuser.me/api') #make a request usin .get
arquivo=json.loads(resposta.text) #loads the response of request as a text in a new variable
```

```python
arquivo
```

Displaying the first and last name:

We acess 'results'(this is a json list, since it's comes before the brackets),[0] to acess the first register(if  this json brought more than a user, we could acess differents users by the index),  inside it find the 'name' (this is a json object, since it'scomes before the curly brackets) and the attribute 'first'. All theses steps brings the first name that is inside our json file

```python
print(arquivo['results'][0]['name']['first'])
print(arquivo['results'][0]['name']['last'])
```

I can also store theses results into variables

```python
first_name=arquivo['results'][0]['name']['first']
last_name=arquivo['results'][0]['name']['last']
print(f'My first name is {first_name} and my last name is {last_name}')
```
