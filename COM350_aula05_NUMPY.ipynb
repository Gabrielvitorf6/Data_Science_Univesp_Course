É importante ressaltar que diferente de uma lista Python, um recorte de um ndarray sofrerá broadcast(A mudança será perpetuada para a variável original(nesse caso = dadosseq), ao contrário do exemplo acima das operações aritméticas) - funciona de maneira específica as views de BDs.

```python
fatiadados = dadosseq[2:4]
```

```python
fatiadados
#Observe que começa a retornar os valores no número exato que passamos o começo(indíce 2) mas ele para uma casa antes no número passado para o fim  fim(indíce 4)
```

```python
fatiadados[1] = fatiadados[1] * 2
```

```python
fatiadados
```

```python
dadosseq
```

```python
fatiadados[:] = 333
```

```python
dadosseq
```

Copiar dados para evitar o broadcast.
No exemplo abaixo observamos como o método copy evade o broadcast ao CRIAR uma cópia invés de USAR uma fatia, como no exemplo acima

```python
fatia2=dadosseq[2:4].copy()
```

```python
fatia2
```

```python
fatia2[:]=111
```

```python
fatia2
```
